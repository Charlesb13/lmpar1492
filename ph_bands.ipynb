```python
bsplotter.plot_brillouin()
```

```python
data = bsplotter.bs_plot_data()
d = data.get('distances')
f = data.get('frequency')

# Calcul dans la direction  - X 

vit1 = 2*np.pi*np.abs((f[0][0][1]-f[0][0][0])*(10**12))/((d[0][1]-d[0][0])*(10**10))
print("La vitesse du son pour la premire branche dans la direction  - X vaut : ", vit1, "m/s.")

# Calcul dans la direction K - 
vit2 = 2*np.pi*np.abs((f[3][1][-1]-f[3][1][-2])*(10**12))/((d[3][-1]-d[3][-2])*(10**10))
print("La vitesse du son pour la premire branche dans la direction K -  vaut : ", vit2, "m/s.")


# Calcul dans la direction  - L
vit3 = 2*np.pi*np.abs((f[4][2][1]-f[4][2][0])*(10**12))/((d[4][1]-d[4][0])*(10**10))
print("La vitesse du son pour la premire branche dans la direction  - L vaut : ", vit3, "m/s.")
```
